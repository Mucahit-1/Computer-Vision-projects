#### TASK: Open the *dog_backpack.jpg* image from the DATA folder and display it in the notebook. Make sure to correct for the RGB order.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
image = cv2.imread("../DATA/dog_backpack.jpg")
```

#### TASK: Flip the image upside down and display it in the notebook.

```python
fliped_img = cv2.flip(rgb_img,0)
plt.imshow(fliped_img)
```

#### TASK: Draw an empty RED rectangle around the dogs face and display the image in the notebook.

```python
cv2.rectangle(rgb_img,pt1=(220,370),pt2=(600,750),color=(255,0,0),thickness=20)
plt.imshow(rgb_img)
```

#### TASK: Draw a BLUE TRIANGLE in the middle of the image. The size and angle is up to you, but it should be a triangle (three sides) in any orientation.

```python
pts = np.array([[430,350],[250,700],[650,700]],dtype=np.int32)
pts3d = pts.reshape((-1,1,2))
```

```python
rgb_img= cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
cv2.polylines(rgb_img,[pts3d],isClosed=True,color=(0,0,255),thickness=20)
plt.imshow(rgb_img)
```

```python
rgb_img= cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
cv2.fillPoly(rgb_img,[pts3d],color=(0,0,255))
plt.imshow(rgb_img)
```
